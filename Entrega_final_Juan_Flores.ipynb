#**Simulacion de datos de dasboard**

```python
import random

def generate_simulated_dashboard_data():
    ventas_actuales = random.randint(100_000_000, 200_000_000)
    ventas_ano_anterior = random.randint(int(ventas_actuales * 0.7), int(ventas_actuales * 0.9))
    variacion_interanual = round(((ventas_actuales - ventas_ano_anterior) / ventas_ano_anterior) * 100, 1)

    margen_bruto_actual = round(random.uniform(35.0, 45.0), 1)
    margen_bruto_anterior = round(random.uniform(37.0, 47.0), 1)

    desvio_presupuesto = round(random.uniform(-10.0, -2.0), 1)

    return f"""
Ventas actuales: {ventas_actuales}
Ventas año anterior: {ventas_ano_anterior}
Variación interanual: {variacion_interanual}%
Margen bruto actual: {margen_bruto_actual}%
Margen bruto anterior: {margen_bruto_anterior}%
Desvío vs presupuesto: {desvio_presupuesto}%
"""

# Generamos los datos del dashboard usando Python para mayor consistencia
datos_dashboard = generate_simulated_dashboard_data()

print("Datos de dashboard simulados (generados por Python):")
print(datos_dashboard)
```

#**Funcion de generacion de prompt de generacion de imagen ejecutiva**

```python
import google.generativeai as genai

def generar(prompt, creatividad=0.7, max_tokens=600):
    try:
        model = genai.GenerativeModel(MODEL_NAME)
        response = model.generate_content(
            prompt,
            generation_config=genai.GenerationConfig(
                temperature=creatividad,
                max_output_tokens=max_tokens
            )
        )
        return response.text
    except Exception as e:
        # Manejo específico para el error de cuota excedida
        if "429" in str(e) and "Quota exceeded" in str(e):
            return "Error: Cuota de API excedida. Por favor, espera un minuto y vuelve a intentarlo.\n\n" + \
                   "Para más información sobre límites de cuota: https://ai.google.dev/gemini-api/docs/rate-limits"
        else:
            return f"Error en generación: {str(e)}"
```

ZERO-SHOT PROMPTING

En este caso el modelo recibe solamente una instruccion donde no se le muestra ningun ejemplo previo.

```python
def narrativa_zero_shot(datos):
    prompt = f"""
    Analiza las siguientes métricas comerciales y explica brevemente qué está ocurriendo
    y cuál podría ser el principal riesgo estratégico.

    Datos:
    {datos}
    """

    return generar(prompt, creatividad=0.3, max_tokens=2000)
```

ONE-SHOT PROMPTING

Para esta funcionalidad se le muestra un ejemplo solo de como debe responder.

```python
def narrativa_one_shot(datos):
    prompt = f"""
    Analiza las siguientes métricas comerciales. Tu respuesta debe ser una explicación ejecutiva clara, usando lenguaje simple y profesional, apta para personas de negocio y no especialistas.
    Responde siguiendo este formato de ejemplo:

    Ejemplo:
    Datos:
    Ventas actuales: 100
    Ventas año anterior: 80
    Margen actual: 30%
    Margen anterior: 35%

    Respuesta esperada:
    Las ventas han mostrado un crecimiento, lo cual es positivo para la empresa. Sin embargo, se observa una ligera caída en el margen bruto, lo que sugiere que podría haber una presión en los costos de los productos vendidos o una estrategia de precios más agresiva. Es crucial investigar la causa de esta disminución en la rentabilidad por unidad para asegurar la sostenibilidad del crecimiento.

    Ahora analiza los siguientes datos y proporciona una respuesta detallada como en el ejemplo:

    {datos}
    """

    return generar(prompt, creatividad=0.3, max_tokens=2000)
```

```python
print("ONE-SHOT:\n")
print(narrativa_one_shot(datos_dashboard))
```

FEW-SHOT PROMPTING

Como tercera alternativa se le muestran varios ejemplos para que aprenda el patrón.

```python
def narrativa_few_shot(datos):
    prompt = f"""
    Analiza las siguientes métricas comerciales y genera una narrativa ejecutiva. Tu respuesta debe ser clara, concisa y fácil de entender tanto para expertos en negocio como para aquellos sin formación específica, siguiendo el patrón de los ejemplos. Asegúrate de explicar las implicaciones de los números.

    Ejemplo 1:
    Ventas actuales: 200
    Ventas anteriores: 150
    Margen actual: 40%
    Margen anterior: 45%
    Resultado:
    La compañía ha experimentado un crecimiento robusto en ventas, lo que indica una buena tracción en el mercado. No obstante, se ha observado una reducción en el margen bruto. Esto sugiere que, si bien se está vendiendo más, la rentabilidad por cada unidad vendida ha disminuido, posiblemente debido a un aumento en los costos de producción o a una estrategia de precios competitiva. Es importante evaluar si este sacrificio de margen es sostenible y está alineado con los objetivos estratégicos a largo plazo.

    Ejemplo 2:
    Ventas actuales: 90
    Ventas anteriores: 120
    Margen actual: 35%
    Margen anterior: 30%
    Resultado:
    Aunque las ventas han disminuido respecto al año anterior, la empresa ha logrado mejorar su margen bruto. Esto indica que, a pesar de vender menos volumen, la rentabilidad por cada venta ha aumentado, lo que podría deberse a una optimización de costos, un enfoque en productos de mayor margen o un ajuste en los precios. Es fundamental analizar si esta mejora en la rentabilidad puede sostenerse o replicarse en un escenario de crecimiento de ventas.

    Ahora analiza los siguientes datos y proporciona una narrativa ejecutiva detallada y comprensible:

    {datos}
    """

    return generar(prompt, creatividad=0.3, max_tokens=2000)
```

```python
print("FEW-SHOT:\n")
print(narrativa_few_shot(datos_dashboard))
```
